```python
# find user agent by writing navigator.useragent in console (press shift+ctrl+J in google chrome)
# https://stackoverflow.com/questions/38489386/python-requests-403-forbidden
# requests  =      2.22.0       
# requests-html =     0.10.0 
import requests

url = 'https://www.cloudtango.org/'
headers={'User-Agent':"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36"}
response = requests.get(url, headers=headers)
response
```

```python
# bs4  = 0.0.1  
from bs4 import BeautifulSoup

b_soup = BeautifulSoup(response.text, 'html.parser')
```

```python
words = ["united states", "meraki", "azure", "managed IT"]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=['name','description','services','partners','HQ'])
df
```

```python
import re
```

```python
row = []
temp= []
body = b_soup.find('tbody')
for a in body.findAll('tr'):
    row.clear()
    company=a.find('td', attrs={'class':'company'})
    name = company.find('img', alt=True)['alt']
    row.append(name)
    description = a.find('td', attrs={'class':None})
    row.append(description.text)
    services = a.find('td', attrs={'class':"hide-for-small"})
    row.append(services.text)
    partners = a.findAll('td', attrs={'class':None})[-1]
    row.append(partners.text)
    location = a.findAll('td', attrs={'class':"hide-for-small"})[-1]
    location = location.find_all(text=True, recursive=False)
    row.append(location)
    temp.clear()
    for l in row:
        temp.append(str(l))
        row = [x.lower() for x in temp]
    if [item for item in row if any((word in item) for word in words)] is not None:
        df = df.append(dict(zip(df.columns, row)), ignore_index=True)
```
